### 4. Creation and Examination of Daily Data

In order to aggregate the data to a daily time frame, we must figure out what function we want to use to aggregate the data. Count will be meaningless since it will just be the number of records for that day (usually 24). Sum would be more interesting for ridership data, but would be difficult to interpret for weather and categorical data. Min and max would be more meaningful for most variables, but would only tell us about one data point in each day. Mean or median is probably the way to go since it provides information about all the intraday observations and is meaningful for all the weather data, including the categorical **weathersit**, which is still on a scale. The mean will even make sense for the true/false fields, such as  **holiday**, and fields that cover timespans larger than one day, such as **season**. So mean it is!

```python
# create the bikeday dataframe using the daily mean and show the stats
bikeday = bike.groupby(bike["dteday"]).mean()
bikeday.describe()
```

The stats table for the daily data looks similar to the stats table for hourly data. However, we now have the usage figures in a daily time series, so we can examine the trend in rider usage over time.

```python
plt.figure(figsize=(15,5))

# average casual riders over time
plt.subplot(1,2,1)
plt.plot(bikeday.casual)
plt.xlabel("day")
plt.ylabel("avg casual")
plt.ylim([0, 300])  # set the y limit to just over the max of casual and registered averages in order to have the same scale

# average registered riders over time
plt.subplot(1,2,2)
plt.plot(bikeday.registered)
plt.xlabel("day")
plt.ylabel("avg registered")
plt.ylim([0, 300])

plt.show()
```

These figures show the clear seasonal trend first found in the hourly data. In addition, there is a clear positive trend in **registered** riders, while there appears to be much a smaller positive in trend. let's dig deeper in the next section using lear regressions to figure out what sort of average usage trends we can find.

### 5. Regression Analysis of Usage Over Time

We want to explain average daily ridership, both **casual** and **registered**, so we will use these as our dependent variable.

Our previous examinations showed that ridership was affected greatly by the day of the week, time of year, and bad weather. Let  us see the effects these have on average daily ridership, and then see if we have any additional trends over time.

In order to make this analysis simpler, let's add a new column that starts at 1, for the first day, and increases by one for each day that passes.

```python
bikeday["day"] = range(1,len(bikeday)+1)
bikeday.head()
```

Now let's see if this variable has any explanitory power over our daily usage figures.

```python
cas_fm = smf.ols(formula='casual ~ day', data=bikeday)
print "**********************\n        CASUAL\n**********************"
print cas_fm.fit().summary()

reg_fm = smf.ols(formula='registered ~ day', data=bikeday)
print "**********************\n      REGISTERED\n**********************"
print reg_fm.fit().summary()
```

The results for **casual** are not great, but **day** is highly significant and has a small positive correlation. However the r-squared is very low, showing that, on its own, **day** only explains a very small portion of change in **casual**.

The results for **registered** are much more promising, but still have an r-squared under 0.5. Again **day** is highly significant, but the coefficent is an order of magnitude larger than that for **casual**.

Let's see if we can improve our regressions by adding in variables for weather.

```python
cas_fm = smf.ols(formula='casual ~ day + weathersit + temp + atemp + hum + windspeed', data=bikeday)
print "**********************\n        CASUAL\n**********************"
print cas_fm.fit().summary()

reg_fm = smf.ols(formula='registered ~ day + weathersit + temp + atemp + hum + windspeed', data=bikeday)
print "**********************\n      REGISTERED\n**********************"
print reg_fm.fit().summary()
```

This greatly improves our regressions, but not all the variables are signficant at the 90% confidence interval, including **temp** for **registered**, and **temp**, **atemp**, and **hum** for **casual**.

We also get a warning about multicollinearity, which makes sense since **temp** and **atemp** are highly correlated, as well as **hum** and **weathersit**.

Given these results let's only remove **atemp** and **hum** and see what we get.

```python
cas_fm = smf.ols(formula='casual ~ day + weathersit + atemp + windspeed', data=bikeday)
print "**********************\n        CASUAL\n**********************"
print cas_fm.fit().summary()

reg_fm = smf.ols(formula='registered ~ day + weathersit + atemp + windspeed', data=bikeday)
print "**********************\n      REGISTERED\n**********************"
print reg_fm.fit().summary()
```

This makes all the variables, except **windspeed** in **casual**, significant to at the 99% confidence interval, and puts our r-squareds at 0.39 and 0.69 for **casual** and **registered** respectively.

Before we move on let's drop **windspeed** from **casual**.

```python
cas_fm = smf.ols(formula='casual ~ day + weathersit + atemp', data=bikeday)
print "**********************\n        CASUAL\n**********************"
print cas_fm.fit().summary()

reg_fm = smf.ols(formula='registered ~ day + weathersit + atemp + windspeed', data=bikeday)
print "**********************\n      REGISTERED\n**********************"
print reg_fm.fit().summary()
```

These are all good improvements, but we should also adjust for the main difference in usage between the two groups, which was that the registered users probably used the service for work. Let's see how adding in **workingday** helps the regression.

```python
cas_fm = smf.ols(formula='casual ~ day + weathersit + atemp + workingday', data=bikeday)
print "**********************\n        CASUAL\n**********************"
print cas_fm.fit().summary()

reg_fm = smf.ols(formula='registered ~ day + weathersit + atemp + windspeed + workingday', data=bikeday)
print "**********************\n      REGISTERED\n**********************"
print reg_fm.fit().summary()
```

This was a great improvement to the explanitory power of the **casual** regression, nearly doubling the r-squared, and a nice bump to the **registered** regression. This gives us two equations that have decent descriptive power and highly significant variables. Let's examine what they mean.

**Casual Regression**

We are able to explain about 2/3rds of the variance for the **casual** average daily ridership with this regression. Going variable by variable we see:
* **day** has a very small effect on ridership, but this can add up over time. All else equal, there are 0.18 new registered customers are added per day, or approximately 9.5 new casual riders per day per year.
* **weathersit** shows a decrease of 12.5 average riders per day as precipitation gets worse. Since the average value for **weathersit** is 1.5, in most instances a good day will add 12.5 average riders
* **atemp** shows that for a 1% increase or decrease in how the temperature feels there will be approximately one additional or less average casual rider.
* **workingday** shows that working days have 32 less average casual riders than non-working days, all else equal

**Registered Regression**

We are able to explain about 78% of the variance for the **registered** average daily ridership with this regression. Going variable by variable we see:
* **day** has a larger effect than for casual riders, with , all else equal, 0.026 new casual customers added per day, or an increase of approximately 65.7 average registered riders per day per year.
* **weathersit** shows a similar negative effect as casual, but is about three times greater at losing 37 average riders per day as precipitation gets worse. This also seems to imply that there will be a sharper spike in registered users on a nicer day than for casual riders.
* **atemp** shows that for a 1% increase or decrease in how the temperature feels there will be approximately 1.5 additional or less average registered riders, in line with the higher impact from **weathersit** compared to casual workers.
* **workingday** shows that working days have 42 more average registered riders than non-working days, all else equal.

### 5. Final Thoughts

This analysis just started to scratch the surface of the different relationships within the bikeshare program data. Here I will discuss a few key trends and how they might be relelvant to the bike share program administrator.

**1. Registered users are using the bike for transportation to and from work**

Hourly and daily trends showed clear spikes in usage during commuting times and during the work week. In addition the registered regression showed a large jump in average participation for work days. There were opposite trends for casual users. Since the registered users make up the majority of customers, it is very iimportant for the bike share program to be able to support the needs of this community:
* Stations should be rebalanced in anticipation of the morning and afternoon rushes if possible. In addition, usage patterns for both registered and casual riders show low participation from around 12 am to 4 am, which may be a good time to do any rebalencing of bikes at different stations.
* Bikes could be designed with more of an eye towards those using the bike for work, including:
    * Allowing priority for registered users during rush hours
    * Improving performance in more weather conditions
    * Adding storage space
 
**2. Bad weather keeps riders off the streets, and good weather leads to spikes in usage**

While this at first seems obvious, this analysis was able to show to what degree this is true. could be used by program administrators to help plan accordingly. 
* Using the predictive regresion models along with weather forecasts would help administrators anticpate when and how much usage would spike and schedule staff accordingly
* Lower usage in winter months could allow for the removal of bikes and/or stations for repairs without providing service outages.

**3. Most of the program growth is coming from registered users**

This is finding is very relevant to shaping the future of the bike share program. Finding that growth in registered users is an order of magnitude larger than the growth in casual users shows that more and more people are considering the bikeshare program an option for their daily commute, but that there is probably a more stable basis of people who use the bikes for pleasure. There are a number of ways that this information could be used by administrators:
* Growth rates could be used to determine future infrastructure needs and inform future plans. For example the regression model could be used to estimate peak usage at some point in the future and the administrator could make sure to have at least that many bikes by that time.
* The lack of growth in casual riders could show a lack of outreach and awareness to this market segment. A marketing campaign could be designed that more directly targets people who use the bikes without registering. Perhaps organizing free bike rides on the weekend. There is also probably a corelation between people using it casually to test out and eventually becoming a registered member.
