### 3. Examination of Hourly Data Graphs

We care about ridership, so let us look at **casual** and **registered** and ignore **cnt** for now, since it is just the other two summed up. Since there are a decent amount of columns in this data, we should focus in on a few ways to examine it.

First lets look at ridership in relation to time:
 * **hr** will give us a continuous look intraday
 * **weekday** will give us a continuous look intraweek
 * **mnth** will give a continuous look intrayear

```python
pd.scatter_matrix(bike[['casual', 'registered', 'hr', 'weekday', 'mnth']], figsize=[10,10], diagonal='kde')
```

_**We can ignore the graphs after the first two rows, since we don't really care about the relationships between hours, days, and years.**_

Looking on the first two rows, we seek that, as suspected, the **casual** and **registered** counts are clumped to the left with a longer tail, as show in the diagonal kde graphs.

**Intraday Data**
* **casual** riders have a nice curve that is lowest in teh early morning, rises gradually to peak at midday, and then gradually lowers back down
* **registered** users have a similar shaped curve, but have large peaks in the morning and evening, with a trough of higher activity in the middle of the day.

This data seams to indicate that during the day, casual riders are using the service in line with general human activity patters and are most active during the midday and early afternoon. On the other hand, the registered users are probably using the service to get to and from work, as seen in the peaks at the beginning and end of the day, with generally more activity in the day then at night.

**Intraweek Data**
* **casual** riders are more active on the weekends than the weekdays, with the highest usage on Saturday
* **registered** riders have the opposite trends, and tend to use more during the week and less on weekends

This upholds the idea that registered riders are using the program more for work, while the casual riders are using it for pleasure on the weekend. Although, both riders have a decent amount of usage on each day.

**Intrayear Data**
* **casual** riders use much less in the winter months, climbing with rates sharply higher in spring and summer, then petering out in summer (but still higher than winter), and climbing back up in fall.
* **registered** ridership showed a steady increase from January onward and only fell slighly when it got cold in November.

These patterns are also consistent, with casual riders being more weather driven and registered riders being more steady, but still somewhat affected by weather. The dip in the summer for casuals and the gentle rise for registered users may also be worth examining.
